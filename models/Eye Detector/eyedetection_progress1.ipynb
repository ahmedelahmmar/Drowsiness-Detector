```python
import cv2
import time

# Load the Haar cascade file for eye detection
eye_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_eye.xml')

# Start capturing video
cap = cv2.VideoCapture(0)

eye_closed_time = 0
eye_closed_start = None

while True:
    ret, frame = cap.read()
    if not ret:
        break

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    eyes = eye_cascade.detectMultiScale(gray, 1.1, 4)

    if len(eyes) == 0:
        if eye_closed_start is None:
            eye_closed_start = time.time()
    else:
        if eye_closed_start is not None:
            eye_closed_time = time.time() - eye_closed_start
            eye_closed_start = None
        else:
            eye_closed_time = 0

    for (x, y, w, h) in eyes:
        center = (x + w // 2, y + h // 2)
        # Make the ellipse more horizontal
        axes = (w // 2, h // 4)  # Adjust the vertical axis for a more oval shape
        angle = 0
        startAngle = 0
        endAngle = 360
        color = (255, 0, 0)
        thickness = 2
        cv2.ellipse(frame, center, axes, angle, startAngle, endAngle, color, thickness)

    cv2.putText(frame, f"Eye Closed Time: {eye_closed_time:.2f} seconds", (10, 30),
                cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2, cv2.LINE_AA)

    cv2.imshow('Frame', frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
